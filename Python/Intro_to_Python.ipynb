# **Functions**

```python
temp = (4 * 9/5) + 32
print(temp)
```

```python
temp = (-6 * 9/5) + 32
print(temp)
```

```python
def c_to_f(c_temp):
  f_temp=(c_temp * (9/5)) + 32
  return f_temp
```

```python
f_temp = c_to_f(4)
print(f_temp)
```

Formatted string literals

```python
print(f'The temperature now is {c_to_f(35)}F')
```

```python
My_Temp_List_C = [0, 10, 20, 30, 40]
```

```python
c_to_f(0)
c_to_f(10)
```
